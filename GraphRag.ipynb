For each cluster (1C), combine into one canonical representative

```python
import re
```

```python
#method to pick canonical representatives
#aviod common nouns [it, the, its]
def get_cannonical_rep(my_cluster):
  #the word with the max length after taking away unallowed characters
  max_word = ""
  #store the aliases as a set (no repeats)
  aliases = set()
  #words that cannot be returned
  not_allow = ["the", "it", "its"]
  for ent in my_cluster:
    #case doesn't matter
    ent = ent.lower()
    #don't allow apostrophes
    # print(f"Original entity: {ent}")
    revised_ent = re.sub(r"’s","",ent)
    # print(f"Revised entity:  {revised_ent}")
    if(revised_ent in not_allow):
      continue
    #if valid, check length and whether it should be max_word, or one of the aliases
    ent_len = len(revised_ent)
    #update the max_word
    if(ent_len) > len(max_word):
      if(max_word != ""):
        aliases.add(max_word)
      max_word = revised_ent
      # print(f"Max word: {max_word}")
    #else add original version to aliases
    else:
      #don't add max_word as an alias of itself
      if revised_ent != max_word:
        aliases.add(revised_ent)
  #return an object containin the max word and all its aliases
  return {"representative":max_word, "aliases": aliases}
```

```python
#store all canonical representatives with aliases
final_rep = []
for my_cluster in preds.get_clusters():
  final_rep.append(get_cannonical_rep(my_cluster))
```
